## TECHNICAL INDICATORS

```python
def add_indicators(data):
    data = data.copy()

    data['sma'] = ta.trend.sma_indicator(data['Close'], window=14)
    data['ema'] = ta.trend.ema_indicator(data['Close'], window=14)
    data['rsi'] = ta.momentum.rsi(data['Close'], window=14)

    data['atr'] = ta.volatility.average_true_range(
        data['High'], data['Low'], data['Close'], window=14
    )

    data['returns'] = data['Close'].pct_change()

    return data
```

## ARIMA PER STOCK

```python
def apply_arima(data):
    close = data['Close'].dropna()

    train_size = int(len(close) * 0.8)
    train = close[:train_size]

    model = ARIMA(train, order=(5,1,0))
    fit = model.fit()

    forecast = fit.forecast(steps=len(close) - train_size)

    data['arima_forecast'] = np.nan
    data.iloc[train_size:, data.columns.get_loc('arima_forecast')] = forecast.values

    return data
```

## GARCH (VOLATILITY)

```python
def apply_garch(data):
    returns = data['returns'].dropna()

    model = arch_model(returns, vol='Garch', p=1, q=1)
    fit = model.fit(disp="off")

    data['volatility'] = np.nan
    data.loc[returns.index, 'volatility'] = fit.conditional_volatility

    return data
```

```python
import pandas as pd
import feedparser
from urllib.parse import quote
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

# =====================================================
# 1. DATE RANGE
# =====================================================
start_date = pd.to_datetime("2023-03-19").date()
end_date = pd.to_datetime("2026-03-19").date()

# =====================================================
# 2. STOCK DATA
# =====================================================
df["Date"] = pd.to_datetime(df["Date"]).dt.date
df = df.sort_values(["Stock", "Date"])

stocks = df["Stock"].unique()

# =====================================================
# 3. SENTIMENT MODEL
# =====================================================
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):
    return analyzer.polarity_scores(str(text))["compound"]

# =====================================================
# 4. SCRAPE NEWS
# =====================================================
news_list = []

for stock in stocks:
    query = quote(f"{stock} stock India")
    url = f"https://news.google.com/rss/search?q={query}"

    feed = feedparser.parse(url)

    for entry in feed.entries:
        if hasattr(entry, "published"):
            try:
                date = pd.to_datetime(entry.published).date()
            except:
                continue

            if start_date <= date <= end_date:
                news_list.append([
                    date,
                    stock,
                    get_sentiment(entry.title)
                ])

news_df = pd.DataFrame(news_list, columns=["Date", "Stock", "Sentiment"])

print("News rows:", len(news_df))

# =====================================================
# 5. AGGREGATE DAILY NEWS
# =====================================================
daily_sentiment = news_df.groupby(["Stock", "Date"]).agg(
    Sentiment=("Sentiment", "mean"),
    News_Count=("Sentiment", "count")
).reset_index()

# =====================================================
# 6. MERGE WITH STOCK DATA
# =====================================================
merged_df = df.merge(
    daily_sentiment,
    on=["Stock", "Date"],
    how="left"
)

# =====================================================
# 7. CORE FIX: REAL MARKET-STYLE SENTIMENT FEATURES
# =====================================================

# keep raw event sentiment
merged_df["Sentiment"] = merged_df["Sentiment"].fillna(0)
merged_df["News_Count"] = merged_df["News_Count"].fillna(0)

# signal strength
merged_df["Sentiment_Impact"] = merged_df["Sentiment"] * merged_df["News_Count"]

# market memory (rolling sentiment)
merged_df["Sentiment_Rolling"] = merged_df.groupby("Stock")["Sentiment"].transform(
    lambda x: x.rolling(7, min_periods=1).mean()
)

# news attention / exposure
merged_df["News_Exposure"] = merged_df.groupby("Stock")["News_Count"].transform(
    lambda x: x.rolling(7, min_periods=1).sum()
)

# =====================================================
# 8. FINANCIAL FEATURES
# =====================================================
merged_df["Returns"] = merged_df.groupby("Stock")["Close"].pct_change()
merged_df["Volatility"] = merged_df.groupby("Stock")["Returns"].rolling(5).std().reset_index(0, drop=True)

# =====================================================
# 9. CLEANUP
# =====================================================
merged_df = merged_df.fillna(0)
merged_df = merged_df.sort_values(["Stock", "Date"]).reset_index(drop=True)

# =====================================================
# 10. OUTPUT CHECK
# =====================================================
print("\nStock rows:", len(df))
print("News rows:", len(news_df))
print("Final rows:", len(merged_df))

print("\nSample:")
print(merged_df)

# =====================================================
# 11. SAVE
# =====================================================
merged_df.to_csv("final_trading_dataset.csv", index=False)
```

```python
def label(row):
    if row['Close'] > row['Open']:
        return 1   # bullish
    elif row['Close'] < row['Open']:
        return -1  # bearish
    return 0

df['Label'] = df.apply(label, axis=1)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

df_ml = df.dropna()

X = df_ml[features]
y = df_ml['Label']

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,
    shuffle=False
)

rf = RandomForestClassifier(
    n_estimators=200,
    random_state=42
)

rf.fit(X_train, y_train)

pred = rf.predict(X_test)
```

```python
import pandas as pd
import numpy as np

# ===============================
# 1. CLEAN + SORT
# ===============================
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Stock', 'Date']).reset_index(drop=True)

# ===============================
# 2. INTRADAY + SWING RETURNS
# ===============================
df['Intraday_Return_%'] = ((df['Close'] - df['Open']) / df['Open']) * 100

N = 5
df['Future_Close'] = df.groupby('Stock')['Close'].shift(-N)
df['Swing_Return_%'] = ((df['Future_Close'] - df['Close']) / df['Close']) * 100

# ===============================
# 3. STRATEGY DATA
# ===============================
strategy = df[[
    'Stock', 'Sector',
    'Intraday_Return_%',
    'Swing_Return_%',
    'Volatility',
    'Sentiment',
    'News_Count' 
]].dropna()

# ===============================
# 4. STOCK SUMMARY
# ===============================
stock_summary = strategy.groupby('Stock').mean(numeric_only=True)

# ===============================
# 5. RELATIVE BENCHMARKS (IMPORTANT FIX)
# ===============================
intraday_median = strategy['Intraday_Return_%'].abs().median()
swing_median = strategy['Swing_Return_%'].abs().median()
vol_median = strategy['Volatility'].median()
sent_median = strategy['Sentiment'].median()
news_median = strategy['News_Count'].median()

# ===============================
# 6. DETAILED INSIGHTS
# ===============================
print("\n📊 FIXED DETAILED STOCK INSIGHTS\n")

for stock in stock_summary.index:

    intraday = stock_summary.loc[stock, 'Intraday_Return_%']
    swing = stock_summary.loc[stock, 'Swing_Return_%']
    risk = stock_summary.loc[stock, 'Volatility']
    sentiment = stock_summary.loc[stock, 'Sentiment']
    news = stock_summary.loc[stock, 'News_Count']

    print("\n" + "="*60)
    print(f"🔹 STOCK: {stock}")
    print("="*60)

    # 📈 Intraday
    if abs(intraday) > intraday_median:
        print("📈 Intraday: Above-average movement (more active stock)")
    else:
        print("📈 Intraday: Below-average movement (stable stock)")

    # 📊 Swing
    if abs(swing) > swing_median:
        print("📊 Swing: Strong directional trend")
    else:
        print("📊 Swing: Weak or sideways movement")

    # ⚡ Risk
    if risk > vol_median:
        print("⚡ Risk: Higher than market average (volatile)")
    else:
        print("⚡ Risk: Lower than market average (stable)")

    # 💬 Sentiment
    if sentiment > sent_median:
        print("💬 Sentiment: More positive than average")
    else:
        print("💬 Sentiment: Neutral or negative relative to market")

    # 📰 News
    if news > news_median:
        print("📰 News: High market attention")
    else:
        print("📰 News: Low market attention")

    # 🧠 Final interpretation (neutral)
    score = (
        (abs(intraday) / (intraday_median + 1e-6)) +
        (abs(swing) / (swing_median + 1e-6)) +
        (risk / (vol_median + 1e-6)) +
        (sentiment)
    )

    if score > 3:
        print("🧠 Overall: Highly dynamic stock (mixed signals, high activity)")
    elif score > 1.5:
        print("🧠 Overall: Moderately active stock")
    else:
        print("🧠 Overall: Stable / low activity stock")
```
